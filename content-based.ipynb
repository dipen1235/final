```python
import requests

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('csv/movies.csv')
credits = pd.read_csv('csv/credits.csv')
```

```python
movies.head(5)
```

```python
credits.head(5)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.columns
```

```python
movies = movies[['movie_id_x','title','overview','genres','keywords','cast','crew']]
```

```python
movies = movies.rename(columns={'movie_id_x': 'movie_id'})
```

```python
movies.columns
```

```python
movies.head(5)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]['crew']
```

```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break

    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(' ',''))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head(5)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(1)
```

```python
new_df.iloc[0]['tags']
```

##We will be using above paragraph to create vectors using count vectorizer

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_df[new_df['title'] == 'Spider-Man'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```
